## Условие: При аэродинамическом торможении в атмосфере планеты температура внутри автоматического спускаемого аппарата повысилась с T = 293 К до  Т = 353 К. Какую часть воздуха необходимо выпустить, чтобы давление внутри аппарата не изменилось? Построить график процесса. 

```python
import matplotlib.pyplot as plt
R = 8.31; T1 = 293; T2 = 353; p = 1.0e5 ; V = 1; mu = 0.029;
N = 100
dT = (T2 - T1) / N
alpha = 1 - T1 / T2
print(" alpha=%6.3f"%alpha)
x=[]; m=[];
x.append(T1) 
m.append((p * V * mu / R) / T1)
for i in range(1, N): 
    t = T1 + i * dT;
    x.append(t)
    m.append((p * V * mu / R) / t) 

plt.figure(figsize=(20, 20))
plt.plot(x, m, 'k-')
plt.xlabel('$T$', fontsize=20)
plt.ylabel('$m$', fontsize=20)
plt.title('Зависимость массы воздуха в аппарате от температуры', fontsize=24)
plt.grid(True)
```
